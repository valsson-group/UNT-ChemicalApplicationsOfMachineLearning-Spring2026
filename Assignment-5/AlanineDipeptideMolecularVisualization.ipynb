```python
print("C7eq Alanine Dipeptide")

import py3Dmol

# view = py3Dmol.view

with open('AlanineDipeptide-Vacuum-C7eq.mol2', 'r') as f:
    pdb_data = f.read()

view = py3Dmol.view()
view.addModel(pdb_data, 'mol2')

view.setStyle("stick")

# view.setStyle({
#             "stick": {"radius": 0.3},     # thicker bonds
#             "sphere": {"scale": 0.25}     # visible atoms
#             })

view.zoomTo()

# view.addLabels("serial")

view.show()
```
