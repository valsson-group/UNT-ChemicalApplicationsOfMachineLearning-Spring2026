## Plotting of data using python


Here we will show a simple example of plotting datasets using python and matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
```

```python
# ploting sin(x)
x = np.linspace(-np.pi,+np.pi,1000)
# print(x)
fx_sin = np.sin(x)
fx_cos = np.cos(x)
plt.plot(x,fx_sin,label='sin(x)')
plt.plot(x,fx_cos,label='cos(x)')
plt.xlim([-np.pi,np.pi])
plt.ylim([-1.2,1.2])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
data = np.loadtxt("Dataset_1D-TimeSeries_Set-1.data")
print(data.shape)
```

```python
print(data)
```

```python
print(data[:,0])
print(data[:,1])
```

```python
plt.plot(data[:,0],data[:,1])
```

```python
time = data[:,0]
variable = data[:,1]
plt.plot(time,variable)
plt.show()
```

```python
# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(variable, bins=50, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Time Series Data')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
import seaborn as sns

plt.hist(
    variable,
    bins=50,
    density=True,
    alpha=0.6,
    edgecolor='black',
    label='Histogram'
)

# KDE plot
sns.kdeplot(
    variable,
    color='red',
    linewidth=2,
    bw_adjust=0.6,
    label='KDE - bw_adjust=0.6'
)

sns.kdeplot(
    variable,
    color='orange',
    linewidth=2,
    bw_adjust=1.0,
    label='KDE - bw_adjust=1.0'
)

sns.kdeplot(
    variable,
    color='purple',
    linewidth=2,
    bw_adjust=2.0,
    label='KDE - bw_adjust=2.0'
)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with KDE Overlay')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

```python
data_2D_TimeSeries = np.loadtxt("Dataset_2D-TimeSeries.data")
print(data_2D_TimeSeries.shape)

time = data_2D_TimeSeries[:,0]
xt = data_2D_TimeSeries[:,1]
yt = data_2D_TimeSeries[:,2]
```

```python

plt.plot(time,xt,label='x(t)',color='red')
plt.legend()
plt.xlabel('t')
plt.show()

plt.plot(time,yt,label='y(t)')
plt.legend()
plt.xlabel('t')
plt.show()
```

```python
plt.plot(xt,yt,'.')
```

```python
sns.kdeplot(
    x=xt,
    y=yt,
    fill=True,
    cmap="viridis",
    levels=30,
    thresh=0
)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D KDE Plot')

plt.show()
```
