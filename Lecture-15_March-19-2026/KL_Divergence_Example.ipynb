## Kullback–Leibler Divergence

```python
# KL divergence of two Gaussians

def KL_Divergence_Gaussians(mu1, sigma1, mu2, sigma2):
  import numpy as np
  import matplotlib.pyplot as plt
  from scipy import stats

  x_min = min(mu1 - 5 * sigma1, mu2 - 5 * sigma2)
  x_max = max(mu1 + 5 * sigma1, mu2 + 5 * sigma2)
  x = np.linspace(x_min, x_max, 10000)

  # Evaluate PDFs
  p = stats.norm.pdf(x, mu1, sigma1)
  q = stats.norm.pdf(x, mu2, sigma2)

  # --- KL Divergence: KL(P || Q) ---
  # Numerical integration: sum p(x) * log(p(x)/q(x)) * dx
  dx = x[1] - x[0]
  kl_numerical_pq = np.sum(p * np.log(p / (q + 1e-10))) * dx
  kl_numerical_qp = np.sum(q * np.log(q / (p + 1e-10))) * dx

  print(f"KL(P || Q):  {kl_numerical_pq:.6f}")
  print(f"KL(Q || P):  {kl_numerical_qp:.6f}")

  # --- Plotting ---
  fig, axes = plt.subplots(1, 3, figsize=(13, 4))

  # 1: PDFs
  ax = axes[0]
  ax.plot(x, p, label=f"P: mu={mu1} sigma={sigma1}", color="steelblue", lw=2)
  ax.plot(x, q, label=f"Q: mu={mu2} sigma={sigma2}", color="tomato", lw=2)
  ax.fill_between(x, p, alpha=0.15, color="steelblue")
  ax.fill_between(x, q, alpha=0.15, color="tomato")
  ax.set_title("Distributions P and Q")
  ax.set_xlabel("x")
  ax.set_ylabel("Density")
  ax.legend()
  ax.grid(True, alpha=0.3)

  # 2: Integrand  p(x) * log(p(x)/q(x))
  integrand = p * np.log(p / (q + 1e-10))
  ax = axes[1]
  ax.plot(x, integrand, color="purple", lw=2)
  ax.fill_between(x, integrand, where=(integrand > 0), alpha=0.2, color="purple", label="positive")
  ax.fill_between(x, integrand, where=(integrand < 0), alpha=0.2, color="orange", label="negative")
  ax.axhline(0, color="black", lw=0.8, ls="--")
  ax.set_title(f"Integrand  p(x)·log(p/q)\nKL(P‖Q) = {kl_numerical_pq:.4f}")
  ax.set_xlabel("x")
  ax.set_ylabel("p(x) · log(p(x)/q(x))")
  ax.legend()
  ax.grid(True, alpha=0.3)

  # 3: Integrand  q(x) * log(q(x)/p(x))
  integrand = q * np.log(q / (p + 1e-10))
  ax = axes[2]
  ax.plot(x, integrand, color="purple", lw=2)
  ax.fill_between(x, integrand, where=(integrand > 0), alpha=0.2, color="purple", label="positive")
  ax.fill_between(x, integrand, where=(integrand < 0), alpha=0.2, color="orange", label="negative")
  ax.axhline(0, color="black", lw=0.8, ls="--")
  ax.set_title(f"Integrand  q(x)·log(q/p)\nKL(Q‖P) = {kl_numerical_qp:.4f}")
  ax.set_xlabel("x")
  ax.set_ylabel("q(x) · log(q(x)/p(x))")
  ax.legend()
  ax.grid(True, alpha=0.3)

  plt.tight_layout()
  plt.show()
```

```python
#@markdown ### Gaussian 1
#@markdown #### - Mean
mu1 = -2 #@param
#@markdown #### - Standard deviation
sigma1 = 1 #@param

#@markdown ### Gaussian 2
#@markdown #### - Mean
mu2 = +2 #@param
#@markdown #### - Standard deviation
sigma2 = 1 #@param

KL_Divergence_Gaussians(mu1, sigma1, mu2, sigma2)

```
