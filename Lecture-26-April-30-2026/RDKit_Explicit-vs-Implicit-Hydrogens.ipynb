```python
from rdkit import Chem
```

```python
Smiles_String="O=C(C)Oc1ccccc1C(=O)O"
```

## Implicit Hydrogens - Default in RDKit

```python
mol=Chem.MolFromSmiles(Smiles_String)
mol
```

```python
for i, atom in enumerate(mol.GetAtoms()):
  print(f"Atom Index: {atom.GetIdx()}, Symbol: {atom.GetSymbol()}, Number of Hydrogens: {atom.GetTotalNumHs()}")
```

## Explict Hydrogens

```python
mol=Chem.MolFromSmiles(Smiles_String)
mol = Chem.AddHs(mol)
mol
```

```python
for i, atom in enumerate(mol.GetAtoms()):
  print(f"Atom Index: {atom.GetIdx()}, Symbol: {atom.GetSymbol()}, Number of Hydrogens: {atom.GetTotalNumHs()}")
```
